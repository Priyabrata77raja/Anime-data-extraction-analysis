```python
#MAKE A COLUMN FOR TIMESTAMP
df.loc[0]["Title"]
```

```python
def extraction_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data += txt[j]

            return data
            
```

```python
df['Total Time'] = df['Title'].apply(extraction_time)
df['Total Time']
```
